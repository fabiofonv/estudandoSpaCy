```python
import spacy
from spacy.matcher import Matcher
from spacy import displacy
import pypdf
```

```python
reader = pypdf.PdfReader("PROVA ESP.PDF")
print(reader.pages[0].extract_text())
```

```python
print(len(reader.pages))
```

```python
nlp = spacy.load("pt_core_news_md")
matcher = Matcher(nlp.vocab)
```

```python
test_doc = nlp(reader.pages[0].extract_text())
# displacy.render(test_doc, style="ent")
for token in test_doc:
    print(f"Token: {token.text}, POS: {token.pos_}, TAG: {token.tag_}")
```

```python
pattern1 = [{"TEXT": {"REGEX": r"^\d{1,3}$"}},
            {"IS_ALPHA": True, "OP": "+"},
            {"IS_PUNCT": True, "OP": "*"},
            {"IS_DIGIT": True, "OP": "*"}]
matcher.add("Q_NUM", [pattern1], greedy="LONGEST")
for text in reader.pages:
    print(text.page_number)
    text = text.extract_text()
    text = text.replace(" ,", ",")
    doc = nlp(text)
    matches = matcher(doc)
    matches.sort(key = lambda x: x[1])
    print (len(matches))
    for match in matches:
        print (match, doc[match[1]:match[2]])
```
